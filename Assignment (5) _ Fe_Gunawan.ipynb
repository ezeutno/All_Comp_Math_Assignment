# Sine Function

## The Tailor Series
This are the equation reagarding to the sin function
$$
\sum\limits_{k=0}^\infty \frac{(-1)^k}{(2k+1)!} {x^{2k+1}} 
$$

Create it into a sin function that are based on 5 (blue), 9 (red), 15(green) iteration.

## Factorial Equation
Since the tailor series equation need a factorial to calcualte the sinus, we must create a factorial equation which is
$$
x! = x(x-1)!
$$
Example factorial of 3
$$
3! = 3\cdot2\cdot1 = 6
$$
The factorial of 3 is 6

```python
def fact(num):
    res = 1
    while (num != 1):
        res = num*res
        num = num-1
    return res
# example of factorial of 5 which is 5*4*3*2*1 = 120
print("Factorial of 5 =",fact(5))
```

## Sum of Exponential Equation
The sum of exponen equation must be calculated since it is has been declare in the equation as in
$$
\sum\limits_{k=0}^n
$$
We must created based on the sin exponential equation which are
$$
\sum\limits_{k=0}^n \frac{(-1)^k}{(2k+1)!} {x^{2k+1}} 
$$
n as in the number of iteration that start from 0

x is degree that are represent in radian of pi

```python
def sin(x,n):
    res = 0;
    for k in range(n):
        res += ((-1)**k/fact(2*k+1))*(x**(2*k+1))
    return res
# for example degree 30 (which is pi*(30/180)) we iterate as many as 100
# to calculate the numpy as accurate as possible we use numpy array provided pi
import numpy as np

print("Numpy Sin of 30 degree : " , np.sin(np.pi/6))
print("Sin from tailor series calculation (100 Iteration) : ",sin(np.pi/6,100))
```
