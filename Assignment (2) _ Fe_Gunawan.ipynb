## Find Minimum Point

$$
f(\bar{x}) = (x_1 - 1)^2 + 100 (x_2 - 1)^2
$$

```python
import matplotlib.pyplot as plt
import sympy as sym
import sympy.plotting.plot as splot
import numpy as np
import math
def steppestDescent(x,f):
    global X
    # print(0,x)
    alpha = 1.0E-3
    i = 0
    while i<=10000:
        print(i,x)
        X.append(x)
        s = grad(x,f)
        x = x - alpha*s
        if math.sqrt(s[0]**2+s[1]**2)<1.0E-6:
            break
        i += 1
def grad(x,f):
    gradx1 = sym.diff(f,x1).subs(x1,x[0])
    gradx2 = sym.diff(f,x2).subs(x2,x[1])
    grad = np.array([gradx1,gradx2])
    return grad
X = []
x1, x2 = sym.symbols('x1 x2')
f = (x1-1)**2 + 100*(x2-1)**2
x = np.array([5,5])
steppestDescent(x,f)
X = np.array(X)
plt.plot(X[:,0],X[:,1])
```
